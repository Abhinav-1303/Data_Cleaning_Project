```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Arcap\Data_Cleaning_Tasks\world_population.csv")
df.head()
```

```python
df_copy = df.copy()
```

```python
# List of numeric columns to validate
numeric_cols = [
    '2022 Population', '2020 Population', '2015 Population',
    '2010 Population', '2000 Population', '1990 Population',
    '1980 Population', '1970 Population',
    'Area (km²)', 'Density (per km²)', 'Growth Rate',
    'World Population Percentage'
]
```

```python
# Convert each column to numeric, coerce invalid entries to NaN
for col in numeric_cols:
    df_copy[col] = pd.to_numeric(df_copy[col], errors='coerce') #If a value is not numeric (like text), it becomes NaN.
```

```python
# Check if any invalid (non-numeric) values were found
invalid_counts = df_copy[numeric_cols].isna().sum()
```

```python
print("Invalid (non-numeric) entries per column:")
print(invalid_counts)
```

```python
print("\nFinal dtypes after conversion:")
print(df_copy.dtypes)
```
